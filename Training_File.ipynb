### Importing Packages        

```python
import pandas as pd
import numpy as np
```

### Load Dataset

```python
dataset = pd.read_csv('data_train.csv')
```

```python
dataset
```

### id as index of Dataframe

```python
dataset.set_index('id',inplace=True)
```

```python
dataset
```

### Count the number of NaN values in each column

```python
for i,j in dataset.isna().sum().iteritems():
    if j != 0:
        print(i,j)
```

### Drop Columns with more NaN

```python
dataset=dataset.drop(columns=['cat6','cat8'])
```

```python
dataset.mean()
```

### Fillna with average values

```python
dataset=dataset.fillna(dataset.mean())
```

```python
dataset.columns
```

### Correlation between columns

```python
dataset.corr()
```

### Check for dupliate rows

```python
for i in dataset.duplicated().items():
    if i==True:
        print (i)
```

### Input and Target

```python
X=dataset.drop('target',axis=1)
Y=dataset.target
```

### Train & Test Split

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, Y_train, Y_test =train_test_split(X, Y, test_size = 0.33, random_state = 5)
```

### Fit Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
model=lr.fit(X_train,Y_train)
```

### Save the model to disk

```python
import pickle
filename = 'finalized_model'
pickle.dump(model, open(filename, 'wb'))
```

### Prediction on Test

```python
Y_pred=lr.predict(X_test)
```

```python
Y_pred
```

### Confusion Matrix

```python
from sklearn import metrics
cm = metrics.confusion_matrix(Y_test, Y_pred)
```

```python
cm
```

### Classification Report

```python
print(metrics.classification_report(Y_test, Y_pred))
```
